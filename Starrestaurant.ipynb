```python
def remove_outliers_iqr(df,columns):
    df_clean = df.copy()
    for col in columns:
        if pd.api.types.is_numeric_dtype(df_clean[col]):
            Q1 = df_clean[col].quantile(0.25)
            Q3 = df_clean[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            initial_count = df_clean.shape[0]
            df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
            removed = initial_count - df_clean.shape[0]
            print(f"Removed {removed} outliers from '{col}'")
    return df_clean
```

```python
columns_to_check = ['Votes', 'Average Cost for two', 'Aggregate rating','Price range']
df = remove_outliers_iqr(df, columns_to_check)


```

```python
franchise_counts = df.groupby('Restaurant Name')['Country'].nunique().reset_index()
franchise_counts.columns = ['Restaurant Name', 'Country Count']

df = pd.merge(df, franchise_counts, on='Restaurant Name', how='left')
```

```python
booking_ratio = df['Has Table booking'].value_counts(normalize=True)
print("\nTable Booking Ratio:\n", booking_ratio)
```

```python
online_delivery_percent = df['Has Online delivery'].value_counts(normalize=True) * 100
print("\nOnline Delivery %:\n", online_delivery_percent)
```

```python
# 📈 Votes Comparison (Delivery vs No Delivery)
delivery_votes = df[df['Has Online delivery'] == 'Yes']['Votes'].sum()
no_delivery_votes = df[df['Has Online delivery'] == 'No']['Votes'].sum()
print(f"\nVotes - Delivery: {delivery_votes}, No Delivery: {no_delivery_votes}, Difference: {delivery_votes - no_delivery_votes}")
```

```python
from collections import Counter
cuisine_list = df['Cuisines'].dropna().apply(lambda x: [c.strip() for c in x.split(',')])
all_cuisines = Counter([item for sublist in cuisine_list for item in sublist])
top_10_cuisines = all_cuisines.most_common(10)
print("\nTop 10 Cuisines:\n", top_10_cuisines)
```

```python
df['Cuisine Count'] = cuisine_list.apply(len)
print("Max cuisines served by a restaurant:", df['Cuisine Count'].max())
print("Min cuisines served by a restaurant:", df['Cuisine Count'].min())
```

```python
city_cuisine = {}
for city in df['City'].unique():
    cuisines_in_city = df[df['City'] == city]['Cuisines'].dropna().str.split(',').sum()
    if cuisines_in_city:
        most_common = pd.Series(cuisines_in_city).str.strip().value_counts().idxmax()
        city_cuisine[city] = most_common
print("\nMost common cuisine per city:\n", city_cuisine)
df['Most Common City Cuisine'] = df['City'].map(city_cuisine)
```

```python
# Define exchange rates to USD
exchange_rates = {
    'Botswana Pula(P)': 0.073,
    'Brazilian Real(R$)': 0.20,
    'Dollar($)': 1.0,
    'Emirati Diram(AED)': 0.27,
    'Indian Rupees(Rs.)': 0.012,
    'NewZealand($)': 0.59,
    'Pounds(å£)': 1.24,
    'Qatari Rial(QR)': 0.27,
    'Rand(R)': 0.054,
    'Sri Lankan Rupee(LKR)': 0.0033,
    'Turkish Lira(TL)': 0.031
}

# Map exchange rates to the dataframe
df['Exchange Rate to USD'] = df['Currency'].map(exchange_rates)

# Convert cost to USD
df['Average Cost for Two (USD)'] = df['Average Cost for two'] * df['Exchange Rate to USD']
```
